```python
class No:
    __valor = None
    __proximo_no = None

    def __init__(self, valor):
        self.__valor = valor

    def valor(self):
        return self.__valor

    def proximo(self):
        return self.__proximo_no

    def apontar_para(self, no):
        self.__proximo_no = no
```

```python
class ListaEncadeada:
    __inicial: No = None

    def __init__(self):
        pass

    def vazia(self):
        return self.__inicial == None

    def inserir_inicio(self, valor):
        no = No(valor)
        if self.vazia():
            self.__inicial = no
        else:
            primeiro_antes = self.__inicial

            no = no
            no.apontar_para(primeiro_antes)

            self.__inicial = no

    def primeiro(self):
        if self.vazia():
            return None
        else:
            return self.__inicial.valor()
        
    def ultimo(self):
        if self.vazia():
            return None
        else:
            no_atual = self.__inicial
            while no_atual.proximo() != None:
                no_atual = no_atual.proximo()

            return no_atual.valor()
        
    def mostrar(self):
        print("--- MOSTRANDO LISTA ---")

        no_atual = self.__inicial
        while no_atual != None:
            print(no_atual.valor())
            no_atual = no_atual.proximo()

        print("-"*8)

    def pesquisar(self, valor):
        if self.vazia():
            return None
        else:
            no_atual = self.__inicial
            while no_atual != None:
                if no_atual.valor() == valor:
                    return no_atual
                no_atual = no_atual.proximo()
            return None
        
    def anterior(self, valor):
        if self.vazia():
            return None
        else:
            no_atual = self.__inicial
            while no_atual.proximo() != None:
                print("proximo", no_atual.proximo().valor())
                if no_atual.proximo().valor() == valor:
                    return no_atual
                no_atual = no_atual.proximo()
            return None
        
    def remover(self, valor):
        alvo = self.pesquisar(valor)
        if not alvo:
            return False
        
        anterior = self.anterior(alvo.valor())
        if not anterior:
            return False
        
        anterior.apontar_para(alvo.proximo())

    def remover_inicio(self):
        primeiro = self.__inicial
        if not primeiro:
            return False
        
        self.__inicial = self.__inicial.proximo()
```

```python
lista = ListaEncadeada()

lista.inserir_inicio("a")
lista.inserir_inicio("b")
lista.inserir_inicio("c")

lista.mostrar()

no_c = lista.pesquisar("a")

print(no_c.valor())

print(lista.anterior("b").valor())

print(lista.remover("b"))

lista.mostrar()

lista.remover_inicio()

lista.inserir_inicio("primeiro!!")

lista.mostrar()

# if lista.vazia():
#     print("Lista vazia!")
# else:
#     print(lista.primeiro())
#     print(lista.ultimo())
```
