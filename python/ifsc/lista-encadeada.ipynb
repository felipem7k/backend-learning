```python
class No:
    __valor = None
    __proximo_no = None

    def __init__(self, valor):
        self.__valor = valor

    def valor(self):
        return self.__valor

    def proximo(self):
        return self.__proximo_no

    def apontar_para(self, no):
        self.__proximo_no = no
```

```python
class ListaEncadeada:
    __inicial: No = None

    def __init__(self):
        pass

    def vazia(self):
        return self.__inicial == None

    def inserir_inicio(self, valor):
        no = No(valor)
        if self.vazia():
            self.__inicial = no
        else:
            primeiro_antes = self.__inicial

            no = no
            no.apontar_para(primeiro_antes)

            self.__inicial = no

    def primeiro(self):
        if self.vazia():
            return None
        else:
            return self.__inicial.valor()
        
    def ultimo(self):
        if self.vazia():
            return None
        else:
            no_atual = self.__inicial
            while no_atual.proximo() != None:
                no_atual = no_atual.proximo()

            return no_atual.valor()
        
    def mostrar(self):
        no_atual = self.__inicial
        while no_atual != None:
            print(no_atual.valor())
            no_atual = no_atual.proximo()
```

```python
lista = ListaEncadeada()

lista.inserir_inicio("a")
lista.inserir_inicio("b")
lista.inserir_inicio("c")

lista.mostrar()

# if lista.vazia():
#     print("Lista vazia!")
# else:
#     print(lista.primeiro())
#     print(lista.ultimo())
```
